# First look at the data

## Import libraries

```python
%matplotlib inline
import numpy as np
from utils.helper import read_data
import utils.processing as processing
import matplotlib.pyplot as plt
FOLDER = "data/"
```

## Load the data

```python
N_STATIC = 3
N_STATIC_F = 3
N_MOVING = 3
data_static = [read_data(FOLDER + "data_static_30_s_" + str(i) + ".csv") for i in range(N_STATIC)]
data_static_f = [read_data(FOLDER + "data_static_1_30_s_" + str(i) + ".csv") for i in range(N_STATIC_F)]
data_moving = [read_data(FOLDER + "data_moving_30_s_" + str(i) + ".csv") for i in range(N_MOVING)]
print(f"Shape static data: {data_static[0].shape}")
print(f"Shape static_f data: {data_static_f[0].shape}")
print(f"Shape moving data: {data_static[0].shape}")
```

### Static data

Look at the phases of static data

```python
phases_static = []
for i in range(N_STATIC):
    phases, abses, _, _ = processing.do_processing(data_static[i]) 
    phases_static.append(np.mean(phases, axis=0))
```

```python
fig, axs = plt.subplots(N_STATIC)
fig.suptitle('Static data')
for i in range(N_STATIC):
    axs[i].plot(phases_static[i])
```

### Static but further

Look as well at the static data, but with Svi sitting further away:)

```python
phases_static_f = []
for i in range(N_STATIC_F):
    phases, abses, _, _ = processing.do_processing(data_static_f[i]) 
    phases_static_f.append(np.mean(phases, axis=0))
```

```python
fig, axs = plt.subplots(N_STATIC_F)
fig.suptitle('Static data but further')
for i in range(N_STATIC_F):
    axs[i].plot(phases_static_f[i])
```

## Moving data 

Data of Svi moving around

```python
phases_moving = []
for i in range(N_MOVING):
    phases, abses, _, _ = processing.do_processing(data_moving[i]) 
    phases_moving.append(np.mean(phases, axis=0))
```

```python
fig, axs = plt.subplots(N_MOVING)
fig.suptitle('Moving data')
for i in range(N_MOVING):
    axs[i].plot(phases_moving[i])
```

### Overall comparison of phases

```python
print("STATIC DATA")
for i in range(N_STATIC):
    max = np.max(phases_static[i])
    min = np.min(phases_static[i])
    print(f"Max: {max:.5} Min: {min:.5}")
    
print("STATIC FURTHER DATA")
for i in range(N_STATIC_F):
    max = np.max(phases_static_f[i])
    min = np.min(phases_static_f[i])
    print(f"Max: {max:.5} Min: {min:.5}")
    
print("MOVING DATA")
for i in range(N_MOVING):
    max = np.max(phases_moving[i])
    min = np.min(phases_moving[i])
    print(f"Max: {max:.5} Min: {min:.5}")
```
